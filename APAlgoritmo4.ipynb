**IMPLEMENTAÇÃO COM TEMPO TOTAL DO ALGORITMO**

```python
import math
from collections.abc import Sequence
from typing import Any, Protocol, TypeVar
import time

class Comparable(Protocol):
    def __lt__(self, other: Any, /) -> bool:
        ...

T = TypeVar("T", bound=Comparable)

def jump_search(arr: Sequence[T], item: T) -> int:
    arr_size = len(arr)
    block_size = int(math.sqrt(arr_size))

    prev = 0
    step = block_size
    while arr[min(step, arr_size) - 1] < item:
        prev = step
        step += block_size
        if prev >= arr_size:
            return -1

    while arr[prev] < item:
        prev += 1
        if prev == min(step, arr_size):
            return -1
    if arr[prev] == item:
        return prev
    return -1

if __name__ == "__main__":
    user_input = input("Insira números separados por vírgula: ").strip()
    array = [int(item) for item in user_input.split(",")]
    x = int(input("Insira o número a ser procurado:\n"))

    start_time = time.time()
    res = jump_search(array, x)
    end_time = time.time()

    if res == -1:
        print("Número não encontrado")
    else:
        print(f"Número {x} no índice {res}")

    execution_time = end_time - start_time
    print(f"Tempo de execução: {execution_time} segundos")
```

**IMPLEMENTAÇÃO DA COMPLEXIDADE DO ALGORITMO**

```python
import math
from collections.abc import Sequence
from typing import Any, Protocol, TypeVar
import time
import random
import matplotlib.pyplot as plt

class Comparable(Protocol):
    def __lt__(self, other: Any, /) -> bool:
        ...

T = TypeVar("T", bound=Comparable)

def jump_search(arr: Sequence[T], item: T) -> int:
    arr_size = len(arr)
    block_size = int(math.sqrt(arr_size))

    prev = 0
    step = block_size
    while arr[min(step, arr_size) - 1] < item:
        prev = step
        step += block_size
        if prev >= arr_size:
            return -1

    while arr[prev] < item:
        prev += 1
        if prev == min(step, arr_size):
            return -1
    if arr[prev] == item:
        return prev
    return -1

def measure_complexity():
    input_sizes = [100, 500, 1000, 5000, 10000]
    execution_times = []

    for size in input_sizes:
        array = [random.randint(1, 100000) for _ in range(size)]
        x = random.randint(1, 100000)

        start_time = time.time()
        jump_search(array, x)
        end_time = time.time()

        execution_time = end_time - start_time
        execution_times.append(execution_time)

    plt.plot(input_sizes, execution_times, marker='o')
    plt.xlabel("Tamanho do vetor")
    plt.ylabel("Tempo de execução")
    plt.title("Complexidade do Algoritmo")
    plt.show()

if __name__ == "__main__":
    measure_complexity()
```

**COMPLEXIDADE**

A complexidade do algoritmo Jump Search no melhor caso é O(√n), porque o algoritmo divide o array em partes de √n e faz uma busca linear no bloco para encontrar o número a ser pesquisado.

No código acima, o bloco de √n são os números que vão ser inseridos separados por vírgula, e em seguida, o algoritmo faz uma busca linear, como mostrado nos dois whiles, para encontrar o número no índice do vetor.
