**1. FATORIAL ITERATIVO**


```python
def fatorial_iterativo(n):
    if n < 0:
        return None
    resultado = 1
    expressao = []
    for i in range(n, 0, -1):
        resultado *= i
        expressao.append(str(i))
        if i != 1:
            expressao.append("*")
    return resultado, expressao

numero = 5
resultado, expressao = fatorial_iterativo(numero)

if resultado is not None:
    print(f'Fatorial de {numero} é {resultado}, o cálculo é: {" ".join(expressao)}')
else:
    print('Insira números positivos')
```

**1. FATORIAL RECURSIVO**


```python
def fatorial_recursivo(n):
    if n == 0:
        return (1, "1")
    else:
        fatorial_anterior, expressao_anterior = fatorial_recursivo(n - 1)
        resultado = n * fatorial_anterior
        expressao = f"{n} * {expressao_anterior}" if n != 1 else str(n)
        return (resultado, expressao)

numero = 6
resultado, expressao = fatorial_recursivo(numero)
print(f'Fatorial de {numero} é {expressao}, o resultado é: {resultado}')
```

**1.1 TEMPO DE EXECUÇÃO ALGORITMO DE FORMA ASSINTÓTICA**

```python
import timeit
import pandas as pd
import matplotlib.pyplot as plt

def fatorial_iterativo(n):
    if n < 0:
        return None
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

def fatorial_recursivo(n):
    if n == 0:
        return 1
    else:
        return n * fatorial_recursivo(n - 1)

valores = list(range(1, 21))

tempos_iterativo = []
for valor in valores:
    tempo = timeit.timeit(lambda: fatorial_iterativo(valor), number=10000)
    tempos_iterativo.append(tempo)

tempos_recursivo = []
for valor in valores:
    tempo = timeit.timeit(lambda: fatorial_recursivo(valor), number=10000)
    tempos_recursivo.append(tempo)

data = {'Valor de n': valores, 'Alg. Iterativo': tempos_iterativo, 'Alg. Recursivo': tempos_recursivo}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(valores, tempos_iterativo, label='Alg. Iterativo')
plt.plot(valores, tempos_recursivo, label='Alg. Recursivo')
plt.xlabel('Valor de n')
plt.ylabel('Tempo')
plt.title('Fatoriais')
plt.grid(True)
plt.show()

print(df)
```

**1.2 O algoritmo recursivo falhará com valores muito grandes de n. A recursão gera uma pilha de chamadas para cada chamada recursiva, e quando n aumenta, o número de chamadas também, causando estouro de pilha. É um comportamento  constante para o algoritmo recursivo e ocorrerá quando n for maior que a capacidade da pilha de chamadas.**


**Já o algoritmo iterativo não falhará devido ao valor de n, por não usar recursão. Uma vez que o algoritmo utiliza apenas um loop simples para calcular o fatorial, torna seu comportamento mais constante, e o algoritmo iterativo é mais adequado do que o algoritmo recursivo para n com valor muito grande Entretanto, para um n com valor baixo, os dois algoritmos possuem desempenho similar.**

```python
import pandas as pd

data = {
    'Algoritmo': ['Iterativo', 'Recursivo'],
    'Complexidade Assintótica': ['O(n)', 'O(n)'],
    'Contagem de Unidades de Tempo': ['Depende do valor de n', 'Depende do valor de n (> iterativo)']
}

df = pd.DataFrame(data)

print(df)
```

Na Contagem de Unidades de Tempo, ambos os algoritmos têm seu tempo de execução dependendo do valor de n. No entanto, o algoritmo recursivo tende a ser mais lento do que o iterativo para n com um valor maior, por causa do custo adicional de chamadas recursivas. Por isso, a contagem de unidades de tempo para o algoritmo recursivo é geralmente maior do que para o algoritmo iterativo, embora os dois tenham a mesma complexidade baseada em n.

**1.4 ANÁLISE DE COMPLEXIDADE ASSINTÓTICA**

A complexidade assintótica do algoritmo iterativo é O(n), sendo n o número que o fatorial é calculado.Logo, o tempo de execução do algoritmo cresce com de maneira linear com base no valor de n. Quanto mais o n aumenta,também aumenta, já que há um loop que executa n vezes, realizando operações em cada iteração.

Já o algoritmo recursivo também tem complexidade assintótica O(n). Mesmo com a recursividade, cada chamada recursiva diminui o valor de n em 1 até chegar a 0. Por isso, o número total de chamadas é igual a n. Como cada chamada realiza uma operação constante, a complexidade do algoritmo ainda é linear.
