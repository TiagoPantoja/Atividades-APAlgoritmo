```python
import matplotlib.pyplot as plt
import numpy as np

n_values = np.arange(1, 101)

f_values = 0.25 * n_values**3 + 0.1 * n_values
g_values = 2 * n_values**2 + 0.5 * n_values + 40

plt.figure(figsize=(8, 6))
plt.plot(n_values, f_values, label='f(n) = 0.25n³ + 0.1n')
plt.plot(n_values, g_values, label='g(n) = 2n² + 0.5n + 40')

plt.title('Complexidades Assintóticas')
plt.xlabel('Número de Entradas')
plt.ylabel('Tempo de Execução')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_values = np.arange(1, 101)

f1_values = n_values
f2_values = 2 * n_values

plt.figure(figsize=(8, 6))
plt.plot(n_values, f1_values, label='f1(n) = n')
plt.plot(n_values, f2_values, label='f2(n) = 2n')

plt.title('Complexidades Assintóticas')
plt.xlabel('Númro de Entradas')
plt.ylabel('Tempo de Execução')

plt.show()
```
