**SCRIPT QUE GERA 10000 PALAVRAS**

```python
import random
import time

with open("palavras.txt", "r", encoding="utf-8") as file:
    palavras_existentes = [linha.strip() for linha in file]
    palavras_geradas = [random.choice(palavras_existentes) for _ in range(10000)]

print("10000 palavras:")
print(palavras_geradas[:10000])

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less_than_pivot = [x for x in arr[1:] if x >= pivot]
        greater_than_pivot = [x for x in arr[1:] if x < pivot]
        return quick_sort(greater_than_pivot) + [pivot] + quick_sort(less_than_pivot)

inicio = time.time()
merge_sort(palavras_geradas.copy())
fim = time.time()
tempo_merge_sort = fim - inicio

inicio = time.time()
palavras_geradas_sorted = quick_sort(palavras_geradas.copy())[::-1]
fim = time.time()
tempo_quick_sort = fim - inicio

print("\nTempo Merge Sort: {:.3f} s".format(tempo_merge_sort))
print("Tempo QuickSort: {:.3f} s".format(tempo_quick_sort))
```

A complexidade assintótica do pior caso para o QuickSort é O(N*2), onde N são as 10000 palavras geradas. O desempenho é quadrático quando o pivô é escolhio de forma incorreta.

A complexidade assintótica do pior caso para o Merge Sort é O(N*log(N)), onde N são as 10000 palavras geradas.

Modificando o Merge Sort pelo HeapSort, e mantendo o QuickSort

```python
import random
import time

with open("palavras.txt", "r", encoding="utf-8") as file:
    palavras_existentes = [linha.strip() for linha in file]
    palavras_geradas = [random.choice(palavras_existentes) for _ in range(10000)]

print("10000 palavras:")
print(palavras_geradas[:10000])

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[i] < arr[left]:
        largest = left

    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heapsort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less_than_pivot = [x for x in arr[1:] if x >= pivot]
        greater_than_pivot = [x for x in arr[1:] if x < pivot]
        return quick_sort(greater_than_pivot) + [pivot] + quick_sort(less_than_pivot)

inicio = time.time()
heapsort(palavras_geradas.copy())
fim = time.time()
tempo_heapsort = fim - inicio

inicio = time.time()
palavras_geradas_sorted = quick_sort(palavras_geradas.copy())[::-1]
fim = time.time()
tempo_quick_sort = fim - inicio

print("\nTempo Heapsort: {:.3f} s".format(tempo_heapsort))
print("Tempo QuickSort: {:.3f} s".format(tempo_quick_sort))
```
